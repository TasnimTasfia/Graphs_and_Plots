```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn
```

```python
df= pd.read_excel('IDF_Step1.xlsx', sheet_name=0)
df.head()
```

# Use Datetime.strftime() Method to extract year

```python
df['Year'] = df['Date'].dt.strftime('%Y')
df.head()
```

```python
df.to_excel("year_data.xlsx")
df1= pd.read_excel('year_data.xlsx',index_col=0,sheet_name=0)
df1.head()
```

```python
res = df1.loc[df1.reset_index().groupby(['Year'])['Precipitation (mm/day) (1  day)'].idxmax()]


print("Result:\n",res)
res.to_excel("max1.xlsx")
```

```python
res2 = df1.loc[df1.reset_index().groupby(['Year'])['Precipitation (mm/day) (2 day)'].idxmax()]
print("Result:\n",res2)
res2.to_excel("max2.xlsx")
```

```python
res3 = df1.loc[df1.reset_index().groupby(['Year'])['Precipitation (mm/day) (3 day)'].idxmax()]
print("Result:\n",res3)
res3.to_excel("max3.xlsx")
```

```python
df= pd.read_excel('Sample IDF Curve - Surma River.xlsx',sheet_name=2)
df.head()
```

```python
plt.rcParams["figure.figsize"] = [6,4]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots(1,1, squeeze= True)

ax.plot(df['Duration (Day)'],df['Intensity (mm) 2 year'],'-D', label='T= 2 year Return Period')
ax.plot(df['Duration (Day)'],df['Intensity (mm) 5 year'],'-D', label='T= 5 year Return Period')
ax.plot(df['Duration (Day)'],df['Intensity (mm) 10 year'],'-D', label='T= 10 year Return Period')
ax.plot(df['Duration (Day)'],df['Intensity (mm) 25 year'],'-D', label='T= 25 year Return Period')


# plt.yticks(np.arange(50,300,25))
plt.title("IDF Curve: Gumbel Distribution", fontweight = 'semibold')
plt.xlabel('Duration (Day)', fontsize=8)
plt.ylabel('Rainfall Intensity (mm)', fontsize=8)
plt.yticks(list(np.arange(100, 450, 50)), alpha=1)
# plt.xticks(list(np.arange(0, 4, .25)), alpha=1)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.grid(visible = True, axis = 'y',  alpha = 0.6)
handles, labels = ax.get_legend_handles_labels()
# fig.legend(handles, labels, loc=(0.1,0.65))
plt.legend()
plt.tight_layout()
for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_linewidth(0.2)
plt.savefig("IDF Curve: Gumbel Distribution_Surma River.png", dpi=500)
```
