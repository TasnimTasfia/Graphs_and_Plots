```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn
```

```python
df= pd.read_excel('DoE_Analysis_Data.xlsx',sheet_name=0)
df.head()
```

```python
df.info('pH')
```

```python
df['pH'] = pd.to_numeric(df['pH'], errors='coerce')
df['DO'] = pd.to_numeric(df['DO'], errors='coerce')
df['TDS'] = pd.to_numeric(df['TDS'], errors='coerce')
df['Turbidity'] = pd.to_numeric(df['Turbidity'], errors='coerce')
df['EC'] = pd.to_numeric(df['EC'], errors='coerce')
```

```python
plt.rcParams["figure.figsize"] = [5.6, 2.8]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots(1,1, squeeze= True)

seaborn.boxplot (x = df['Month'], y = df['pH'],
                ax = ax, color = 'green')

plt.title("pH", fontweight = 'semibold')
plt.xlabel('')
plt.ylabel('pH', fontsize=8)
plt.yticks(list(np.arange(6.5, 9, .5)), alpha=1)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.grid(visible = True, axis = 'y',  alpha = 0.6)
plt.tight_layout()
for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_linewidth(0.2)
plt.savefig("pH_H.png", dpi=500)
```
